```python
import pandas as pd
import nltk
from textblob import TextBlob
import seaborn as sns
```

# Earth Day

```python
# Earth Day - Analysis
file_earth_day = open('../tweets-earth-day.txt', 'r')
lines = file_earth_day.readlines()

earth_day_analyzed = []

for tweet in lines:
  analyzed = TextBlob(tweet)
  earth_day_analyzed.append([tweet, analyzed.sentiment.polarity, analyzed.sentiment.subjectivity])

earth_day_df = pd.DataFrame(earth_day_analyzed, columns=['text', 'polarity', 'subjectivity'])

earth_day_df
```

```python
# Earth Day - Polarity graph
sns.histplot(earth_day_df['polarity'])
```

```python
# Earth Day - Subjectivity graph
sns.histplot(earth_day_df['subjectivity'])
```

# Words analysis

```python
from nltk import download
download('stopwords')
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
nltk.download('punkt')
```

```python
stop_words = stopwords.words('english')

earth_day_data = ' '.join(earth_day_df['text'].tolist()).lower()
biden_data = ' '.join(biden_df['text'].tolist()).lower()
bitcoin_data = ' '.join(bitcoin_df['text'].tolist()).lower()

def remove_stop_words(sentence):
    return ' '.join([word for word in sentence if word not in stop_words and len(word) > 2])

earth_day_data_clean = remove_stop_words(word_tokenize(earth_day_data))
biden_data_clean = remove_stop_words(word_tokenize(biden_data))
bitcoin_data_clean = remove_stop_words(word_tokenize(bitcoin_data))

earth_day_tokens = word_tokenize(earth_day_data_clean)
biden_tokens = word_tokenize(biden_data_clean)
bitcoin_tokens = word_tokenize(bitcoin_data_clean)

fdist_earth_day = FreqDist(earth_day_tokens)
fdist_biden = FreqDist(biden_tokens)
fdist_bitcoin = FreqDist(bitcoin_tokens)
```
